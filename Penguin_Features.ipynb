Scientists have gathered data on three different species of penguins: Adelie, Gentoo, and Chinstrap. Now they want to determine which features can be represented by a normal distribution and whether there is any skew.

Dataset: https://static.junilearning.com/python_data_science/penguins.csv

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import math
```

Start by loading the data into a DataFrame.


```python
penguinDF = pd.read_csv("https://static.junilearning.com/python_data_science/penguins.csv")
display(penguinDF)
```

Now plot a histogram for each quantitative column. Make sure to include a KDE curve. It might be helpful to write a function that creates a histogram for every column. <br> Do any of the distributions look normal? Is there any skew? What does this tell us? <br> What might be causing problems (if there are any)?


```python
def createPlot (df,colName):
  sns.histplot(df,x=colName,kde=True)
  plt.show()

print(createPlot(penguinDF,"bill_length_mm"))
#Do any look normal? The bill_depth looks normal. Body_mass is right skewed.
```

```python
def createPlots(df):
  createPlot(df,"bill_length_mm")
  createPlot(df,"bill_depth_mm")
  createPlot(df,"flipper_length_mm")
  createPlot(df,"body_mass_g")

print(createPlots(penguinDF))
```

Let's create a seperate DataFrame for each species of penguin and regraph the columns. <br> Now do any of the distributions look normal? Do any of them have any skew? Why or why not?
<br>

```python
adelieDF = penguinDF[penguinDF["species"] == "Adelie"]
gentooDF = penguinDF[penguinDF["species"] == "Gentoo"]
chinstrapDF = penguinDF[penguinDF["species"] == "Chinstrap"]
```

```python
createPlots(adelieDF)
```

```python
createPlots(gentooDF)
```

```python
createPlots(chinstrapDF)
```

Finally, lets plot the same data one last time, but isolate the males and females from each other. 
<br> First plot all the female statistics, and then plot all the male statistics.
<br> What differences does this cause? Do any of the distributions look normal? Do any of them have any skew? Why or why not? What could improve our results further?
<br>


```python
# For Females
adelieFDF = adelieDF[adelieDF["sex"] == "FEMALE"]
gentooFDF = gentooDF[gentooDF["sex"] == "FEMALE"]
chinstrapFDF = chinstrapDF[chinstrapDF["sex"] == "FEMALE"]

print("Adelie Female")
createPlots(adelieFDF)
print("Gentoo Female")
createPlots(gentooFDF)
print("Chinstrap Female")
createPlots(chinstrapFDF)

#When filtered by species and gender, the data looks more normal. What could improve our results is if we were given a bigger population.
```

```python
# For Males

adelieMDF = adelieDF[adelieDF["sex"] == "MALE"]
gentooMDF = gentooDF[gentooDF["sex"] == "MALE"]
chinstrapMDF = chinstrapDF[chinstrapDF["sex"] == "MALE"]

print("Adelie Male")
createPlots(adelieMDF)
print("Gentoo Male")
createPlots(gentooMDF)
print("Chinstrap Male")
createPlots(chinstrapMDF)
```
